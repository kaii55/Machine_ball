```python
import pandas as pd 
import numpy as np

df_wwcup = pd.read_excel("WWCUP_shots.xlsx")
df_wwcup['league_trophy'] = 'WWCUP'
print(df_wwcup.shape)
df_laliga = pd.read_excel("Laligafinal_shots.xlsx")
df_laliga['league_trophy'] = 'Laliga'
print(df_laliga.shape)

df_epl = pd.read_excel("EPL_shots.xlsx")
df_epl['league_trophy'] = 'EPL'
print(df_epl.shape)
df_fawsl = pd.read_excel("FAWSL_shots.xlsx")
df_fawsl['league_trophy'] = 'FAWSL'
print(df_fawsl.shape)
df_fwcup = pd.read_excel("FWCUP_shots.xlsx")
df_fwcup['league_trophy'] = 'FWCUP'
print(df_fwcup.shape)

df_isl = pd.read_excel("ISL_shots.xlsx")
df_isl['league_trophy'] = 'ISL'
print(df_isl.shape)
df_nwsl = pd.read_excel("NWSL_shots.xlsx")
df_nwsl['league_trophy'] = 'NWSL'
print(df_nwsl.shape)
df_ucl = pd.read_excel("UCLfinal_shots.xlsx")
df_ucl['league_trophy'] = 'UCL'
print(df_ucl.shape)


df_ueuro = pd.read_excel("UEURO_shots.xlsx")
df_ueuro['league_trophy'] = 'UEURO'
print(df_ueuro.shape)
df_uweuro = pd.read_excel("UWEURO_shots.xlsx")
df_uweuro['league_trophy'] = 'UWEURO'
print(df_uweuro.shape)


frames = [df_wwcup, df_laliga, df_epl, df_fawsl, df_fwcup, 
         df_isl, df_nwsl, df_ucl, df_ueuro, df_uweuro]

df = pd.concat(frames)

print(df.shape)
print(df.columns)
```

```python
df['league_trophy'].value_counts()
```

```python
import ast

for i in range(0, df.shape[0]-1):
    try:
        df['shot_freeze_frame'].iloc[i] = ast.literal_eval(df['shot_freeze_frame'].iloc[i])
    except:
        None
```

```python
def dict_to_dataframe(a):
    dataframe = df[a].apply(pd.Series)
    dataframe.columns = [str(a) + "_" + str(col) for col in dataframe.columns]
    return dataframe

df_sff = dict_to_dataframe('shot_freeze_frame')
print(df_sff.shape)
```

```python
df_sff.head()
```

```python
from pandas import isnull
df_sff=df_sff.applymap(lambda x: {'location': [0, 0], 'position': {'name': 'Center Defensive Midfield'}} if isnull(x) else x)
```

```python
def gk_pos(col):
    pos_x = []
    pos_y = []
    for i in range(0, df.shape[0]):
        try:
            if df_sff[col].values[i]['position']['name'] == 'Goalkeeper':
                x = df_sff[col].values[i]['location'][0]
                y = df_sff[col].values[i]['location'][1]
                pos_x.append(x)
                pos_y.append(y)
            else:
                pos_x.append(0)
                pos_y.append(0)
        except:
            None
    return pos_x, pos_y
        
```

```python
a0_x, a0_y = gk_pos('shot_freeze_frame_0')
df_sff['gk_pos_0_x'] = pd.DataFrame(a0_x)
df_sff['gk_pos_0_y'] = pd.DataFrame(a0_y)


a1_x, a1_y = gk_pos('shot_freeze_frame_1')
df_sff['gk_pos_1_x'] = pd.DataFrame(a1_x)
df_sff['gk_pos_1_y'] = pd.DataFrame(a1_y)


a2_x, a2_y = gk_pos('shot_freeze_frame_2')
df_sff['gk_pos_2_x'] = pd.DataFrame(a2_x)
df_sff['gk_pos_2_y'] = pd.DataFrame(a2_y)


a3_x, a3_y = gk_pos('shot_freeze_frame_3')
df_sff['gk_pos_3_x'] = pd.DataFrame(a3_x)
df_sff['gk_pos_3_y'] = pd.DataFrame(a3_y)


a4_x, a4_y = gk_pos('shot_freeze_frame_4')
df_sff['gk_pos_4_x'] = pd.DataFrame(a4_x)
df_sff['gk_pos_4_y'] = pd.DataFrame(a4_y)


a5_x, a5_y = gk_pos('shot_freeze_frame_1')
df_sff['gk_pos_5_x'] = pd.DataFrame(a5_x)
df_sff['gk_pos_5_y'] = pd.DataFrame(a5_y)


a6_x, a6_y = gk_pos('shot_freeze_frame_6')
df_sff['gk_pos_6_x'] = pd.DataFrame(a6_x)
df_sff['gk_pos_6_y'] = pd.DataFrame(a6_y)


a7_x, a7_y = gk_pos('shot_freeze_frame_7')
df_sff['gk_pos_7_x'] = pd.DataFrame(a7_x)
df_sff['gk_pos_7_y'] = pd.DataFrame(a7_y)


a8_x, a8_y = gk_pos('shot_freeze_frame_8')
df_sff['gk_pos_8_x'] = pd.DataFrame(a8_x)
df_sff['gk_pos_8_y'] = pd.DataFrame(a8_y)


a9_x, a9_y = gk_pos('shot_freeze_frame_9')
df_sff['gk_pos_9_x'] = pd.DataFrame(a9_x)
df_sff['gk_pos_9_y'] = pd.DataFrame(a9_y)


a10_x, a10_y = gk_pos('shot_freeze_frame_10')
df_sff['gk_pos_10_x'] = pd.DataFrame(a10_x)
df_sff['gk_pos_10_y'] = pd.DataFrame(a10_y)


a11_x, a11_y = gk_pos('shot_freeze_frame_11')
df_sff['gk_pos_11_x'] = pd.DataFrame(a11_x)
df_sff['gk_pos_11_y'] = pd.DataFrame(a11_y)

a12_x, a12_y = gk_pos('shot_freeze_frame_12')
df_sff['gk_pos_12_x'] = pd.DataFrame(a12_x)
df_sff['gk_pos_12_y'] = pd.DataFrame(a12_y)


a13_x, a13_y = gk_pos('shot_freeze_frame_13')
df_sff['gk_pos_13_x'] = pd.DataFrame(a13_x)
df_sff['gk_pos_13_y'] = pd.DataFrame(a13_y)


a14_x, a14_y = gk_pos('shot_freeze_frame_14')
df_sff['gk_pos_14_x'] = pd.DataFrame(a14_x)
df_sff['gk_pos_14_y'] = pd.DataFrame(a14_y)


a15_x, a15_y = gk_pos('shot_freeze_frame_15')
df_sff['gk_pos_15_x'] = pd.DataFrame(a15_x)
df_sff['gk_pos_15_y'] = pd.DataFrame(a15_y)

a16_x, a16_y = gk_pos('shot_freeze_frame_16')
df_sff['gk_pos_16_x'] = pd.DataFrame(a16_x)
df_sff['gk_pos_16_y'] = pd.DataFrame(a16_y)


a17_x, a17_y = gk_pos('shot_freeze_frame_17')
df_sff['gk_pos_17_x'] = pd.DataFrame(a17_x)
df_sff['gk_pos_17_y'] = pd.DataFrame(a17_y)


a18_x, a18_y = gk_pos('shot_freeze_frame_18')
df_sff['gk_pos_18_x'] = pd.DataFrame(a18_x)
df_sff['gk_pos_18_y'] = pd.DataFrame(a18_y)


a19_x, a19_y = gk_pos('shot_freeze_frame_19')
df_sff['gk_pos_19_x'] = pd.DataFrame(a19_x)
df_sff['gk_pos_19_y'] = pd.DataFrame(a19_y)

a20_x, a20_y = gk_pos('shot_freeze_frame_20')
df_sff['gk_pos_20_x'] = pd.DataFrame(a20_x)
df_sff['gk_pos_20_y'] = pd.DataFrame(a20_y)


```

```python
df_sff = df_sff[['gk_pos_x', 'gk_pos_y']]
```

```python
df_sff['gk_pos_x'] = df_sff['gk_pos_x'].replace(0.0, 119.1)
df_sff['gk_pos_y'] = df_sff['gk_pos_y'].replace(0.0, 37.8)
```

```python
df_shots = pd.concat([df, df_sff], axis = 1)
```

```python
df_shots.columns
```

```python
df_shots['gk_distance'] = np.sqrt(((df_shots['location_x'] - df_shots['gk_pos_x'])**2) + ((df_shots['location_y'] - df_shots['gk_pos_y'])**2))
```

```python
df_shots['gk_distance'] = df_shots['gk_distance'].round(2)
df_shots.head()
```

```python
df_shots['gk_distance'] = df_shots['gk_distance'].fillna(0)
```
